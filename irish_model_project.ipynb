```python
import pandas as pd                                                 #to import an files
import numpy as np                                                  #to use an array element
import matplotlib.pyplot as plt                                     #for mathematical calculations
from sklearn.neighbors import KNeighborsClassifier                  #to creat relations between functions
from sklearn import preprocessing                                   #to predict the values
from sklearn.model_selection import train_test_split                #for spliting the data into test train and split
from sklearn.metrics import accuracy_score
iris=pd.read_csv("/content/iris.csv")                             #read the csv file
```

```python
x=iris.iloc[:,:4]                               #here we reading only first 4 columns
y=iris.iloc[:,-1]                               # displaying only last column
y
```

```python
x
```

```python
x=preprocessing.StandardScaler().fit_transform(x)      #converting character into the array to predict the values(preprocessing and fitting)
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=34) #dividing an data into test train and split
x_train.shape,x_test.shape
```

```python
knmodel=KNeighborsClassifier(n_neighbors=3)                         #kneighbours will compair the other values and predect the unknown values
knmodel.fit(x_train,y_train)                                        #fiting the data
```

```python
y_predict=knmodel.predict(x_test)             #predicting values
y_predict
```

```python
 from sklearn.metrics import accuracy_score                        #importing the accuracy of the project
acc=accuracy_score(y_test.values,y_predict)
acc
```
